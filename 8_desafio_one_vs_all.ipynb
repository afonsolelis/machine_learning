# Resolução

A classificação de dados é uma tarefa comum em aprendizado de máquina, onde o objetivo é prever a classe de um dado com base em suas características. Existem várias estratégias para abordar tarefas de classificação, e duas delas são conhecidas como "One vs One" e "One vs Rest".

Neste desafio, exploraremos essas estratégias usando o conjunto de dados Wine disponível na biblioteca Scikit-learn. O conjunto de dados Wine contém informações sobre diferentes características de amostras de vinhos, como teor alcoólico, acidez, entre outros. O objetivo é classificar as amostras em três classes diferentes, representando diferentes tipos de vinhos.

Para realizar essa tarefa, utilizaremos classificadores de Árvore de Decisão, que são modelos de aprendizado de máquina capazes de aprender padrões complexos nos dados e tomar decisões de classificação com base nesses padrões.

Primeiro, carregaremos o conjunto de dados, dividiremos os dados em conjuntos de treinamento e teste e, em seguida, implementaremos e treinaremos classificadores de Árvore de Decisão para cada uma das estratégias: "One vs One" e "One vs Rest". Em seguida, avaliaremos o desempenho desses classificadores e compararemos suas acurácias.

Além disso, apresentaremos visualizações que ajudarão a entender melhor as diferenças de desempenho entre essas duas estratégias. Ao final deste desafio, você terá uma compreensão sólida das estratégias "One vs One" e "One vs Rest" em tarefas de classificação e como implementá-las usando Árvores de Decisão.

Vamos começar explorando e classificando vinhos com essas estratégias de classificação!

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split  # Importe train_test_split para dividir o conjunto de dados

# Passo 1: Carregue o conjunto de dados
wine = load_wine()

# Passo 2: Divida o dataset em conjunto de treinamento e de teste
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=42)

# Passo 3: Implemente um classificador de árvore de decisão para cada estratégia
clf_ovo = DecisionTreeClassifier(criterion='entropy', splitter='best')
clf_ovr = DecisionTreeClassifier(criterion='entropy', splitter='best')

# Passo 4: Treine os classificadores nos conjuntos de treinamento
clf_ovo.fit(X_train, y_train)
clf_ovr.fit(X_train, y_train)

# Passo 5: Avalie os classificadores nos conjuntos de teste
y_pred_ovo = clf_ovo.predict(X_test)
y_pred_ovr = clf_ovr.predict(X_test)

# Passo 6: Compare os resultados dos classificadores
print(f'Acurácia One vs One: {accuracy_score(y_test, y_pred_ovo):.3f}')
print(f'Acurácia One vs Rest: {accuracy_score(y_test, y_pred_ovr):.3f}')

# Visualizador: Vamos criar um gráfico de barras para visualizar as acurácias
import matplotlib.pyplot as plt

labels = ['One vs One', 'One vs Rest']
accuracies = [accuracy_score(y_test, y_pred_ovo), accuracy_score(y_test, y_pred_ovr)]

plt.figure(figsize=(8, 6))
plt.bar(labels, accuracies, color=['lightblue', 'lightgreen'])
plt.xlabel('Estratégia de Classificação')
plt.ylabel('Acurácia')
plt.title('Acurácia da Estratégia One vs One vs One vs Rest')
plt.ylim(0.85, 1.0)  # Defina o limite do eixo y para uma melhor visualização
plt.show()
```
