# Feature Scaling, Robust Scaling e Normalização de Dados

**Feature Scaling**: Uma Prática Importante na Preparação de Dados para Machine Learning

Em Machine Learning, o sucesso de um modelo muitas vezes depende da qualidade e do pré-processamento adequado dos dados de entrada. Uma etapa fundamental nesse processo é o _feature scaling_, que se refere à normalização das características (ou _features_) de um conjunto de dados. A normalização das características é crucial para garantir que o desempenho dos algoritmos de Machine Learning não seja afetado por diferenças nas escalas das características.

**Por que é necessário o Feature Scaling?**

Vamos entender a importância do _feature scaling_ com um exemplo usando o conjunto de dados de câncer de mama disponível no Scikit-Learn. Este conjunto de dados contém informações sobre diversas características das células tumorais, e o objetivo é prever se o tumor é benigno ou maligno.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

Vamos dar uma olhada em algumas características deste conjunto de dados:

* Raio médio das células (mean radius): Valores variando de 6 a 28.
* Textura média das células (mean texture): Valores variando de 9 a 39.
* Área média do tumor (mean area): Valores variando de 143 a 2

Aqui está uma representação das escalas das três características mencionadas:
u98.png)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Criar um DataFrame de exemplo com as características
data = {
    'Característica': ['Raio Médio', 'Textura Média', 'Área Média'],
    'Valor Mínimo': [6, 9, 143],
    'Valor Máximo': [28, 39, 2501]
}

df = pd.DataFrame(data)

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df['Característica'], df['Valor Mínimo'], label='Valor Mínimo', color='skyblue', alpha=0.7)
plt.bar(df['Característica'], df['Valor Máximo'], label='Valor Máximo', color='orange', alpha=0.7)

plt.xlabel('Característica')
plt.ylabel('Valor')
plt.title('Escalas das Características')
plt.legend()
plt.show()
```

Como podemos ver, essas características têm escalas muito diferentes. Se aplicarmos algoritmos de Machine Learning diretamente a esses dados sem normalizá-los, os algoritmos que dependem de cálculos de distância, como k-Means ou Support Vector Machines (SVM), serão fortemente influenciados pelas características que têm valores maiores. Isso pode levar a um desempenho subótimo do modelo.

**Métodos Comuns de Feature Scaling**

Existem dois métodos comuns de _feature scaling_: **MinMax Scaling** e **Standardization (Z-score scaling)**.

1. **MinMax Scaling**: Este método dimensiona as características para um intervalo específico, geralmente entre 0 e 1. O valor mínimo da característica é transformado em 0, o valor máximo é transformado em 1 e os valores intermediários são dimensionados proporcionalmente. A fórmula geral é:
    
    Xscaled=X−XminXmax−XminX_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}Xscaled​=Xmax​−Xmin​X−Xmin​​
    
    Isso garante que todas as características estejam no mesmo intervalo.
    
2. **Standardization (Z-score scaling)**: Este método dimensiona as características de forma que elas tenham média zero e desvio padrão um. A fórmula geral é:
    
    Xscaled=X−μσX_{scaled} = \frac{X - \mu}{\sigma}Xscaled​=σX−μ​
    
    Onde μ\muμ é a média da característica e σ\sigmaσ é o desvio padrão. Isso torna as características comparáveis e mantém a distribuição das características aproximadamente normal.
    

**Exemplo Prático de Feature Scaling:**

Vamos aplicar o _feature scaling_ ao conjunto de dados de câncer de mama usando o MinMax Scaling como exemplo:

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Carregar o conjunto de dados de câncer de mama
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Aplicar o MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Exibir as estatísticas descritivas antes e depois do escalonamento
print("Estatísticas descritivas antes do escalonamento:")
print(df.describe())

print("\nEstatísticas descritivas após o escalonamento:")
print(df_scaled.describe())

# Exibir um gráfico de boxplot antes e depois do escalonamento
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df.values)
plt.title("Boxplot Antes do Escalonamento")
plt.xticks(range(1, len(df.columns) + 1), df.columns, rotation=90)

plt.subplot(1, 2, 2)
plt.boxplot(df_scaled.values)
plt.title("Boxplot Após o Escalonamento")
plt.xticks(range(1, len(df.columns) + 1), df.columns, rotation=90)

plt.tight_layout()
plt.show()
```

Agora, as características foram dimensionadas para o intervalo [0, 1], tornando-as comparáveis e adequadas para algoritmos de Machine Learning que dependem de distâncias ou cálculos de similaridade. Este código carrega o conjunto de dados de câncer de mama, aplica o MinMaxScaler, exibe as estatísticas descritivas antes e depois do escalonamento e mostra um gráfico de boxplot comparando as distribuições das características antes e depois do escalonamento.

O _feature scaling_ é uma etapa essencial no pré-processamento de dados para Machine Learning. Ele ajuda a garantir que as características não influenciem negativamente o desempenho dos algoritmos devido a diferenças nas escalas. Portanto, ao lidar com conjuntos de dados heterogêneos, lembre-se de aplicar o _feature scaling_ apropriado para obter resultados mais precisos e confiáveis.

## Entendendo a Padronização e Normalização de Dados

Ao trabalhar com conjuntos de dados, é fundamental preparar os dados de maneira adequada para garantir que os modelos de machine learning funcionem bem. A padronização e a normalização são duas técnicas essenciais para alcançar esse objetivo.

Padronização (Standardizing)
----------------------------

A padronização envolve a transformação dos dados de forma que tenham uma média zero e um desvio padrão igual a um. Isso é útil quando as características têm escalas diferentes e é importante que todas tenham a mesma influência no modelo. Vamos ilustrar isso usando o conjunto de dados de câncer de mama do scikit-learn.

Primeiro, vamos carregar o conjunto de dados e observar as distribuições das características antes da padronização:ização:

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o conjunto de dados de câncer de mama
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Exibir histogramas das características antes da padronização
df.hist(figsize=(12, 10))
plt.suptitle("Histogramas das Características Antes da Padronização", y=1.02)
plt.tight_layout()
plt.show()
```

Os histogramas mostram que as características têm diferentes escalas, o que pode afetar a performance dos modelos. Agora, vamos padronizar os dados:

```python
from sklearn.preprocessing import StandardScaler

# Aplicar a padronização
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Exibir histogramas das características após a padronização
df_standardized.hist(figsize=(12, 10))
plt.suptitle("Histogramas das Características Após a Padronização", y=1.02)
plt.tight_layout()
plt.show()
```

Após a padronização, todas as características têm média zero e desvio padrão igual a um, o que ajuda a evitar que características com escalas maiores dominem o modelo.

Normalização (Normalizing)
--------------------------

A normalização é outra técnica importante que dimensiona os dados em um intervalo específico, geralmente de 0 a 1. Isso é útil quando os dados possuem intervalos diferentes e queremos que todos estejam na mesma faixa.

Vamos aplicar a normalização ao mesmo conjunto de dados:


```python
from sklearn.preprocessing import MinMaxScaler

# Aplicar a normalização
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Exibir histogramas das características após a normalização
df_normalized.hist(figsize=(12, 10))
plt.suptitle("Histogramas das Características Após a Normalização", y=1.02)
plt.tight_layout()
plt.show()
```

Agora, todas as características estão no intervalo de 0 a 1. Isso pode ser útil quando queremos que todas as características tenham a mesma escala relativa, , a padronização e a normalização são técnicas essenciais para garantir que as características dos dados estejam na mesma escala e sejam apropriadas para a construção de modelos de machine learning. A escolha entre essas técnicas depende do problema em questão e das características dos dados.

## Escala Robusta (Robust Scaling) na Prática com Scikit Learn

Ao lidar com conjuntos de dados reais, frequentemente encontramos valores extremos ou outliers que podem distorcer nossos resultados quando escalamos os dados. A escala robusta é uma técnica que lida bem com esses outliers, tornando-a uma escolha ideal em muitos cenários.

O que é Escala Robusta?
-----------------------

A escala robusta, também conhecida como robust scaling, é uma técnica de escalonamento de dados que é menos sensível a valores extremos (outliers) do que a padronização tradicional. Ela utiliza estatísticas resistentes a outliers para dimensionar os dados, tornando-a uma opção sólida quando nossos conjuntos de dados podem conter valores discrepantes.

Conjunto de Dados Iris
----------------------

Para ilustrar a escala robusta em ação, usaremos o conjunto de dados Iris, que contém medidas de diferentes espécies de flores. Vamos começar carregando o conjunto de dados e dando uma olhada em algumas de suas características.


```python
import pandas as pd
from sklearn.datasets import load_iris

# Carregar o conjunto de dados Iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Exibir as primeiras linhas do conjunto de dados
print(df.head())
```

Aplicando a Escala Robusta
--------------------------

Agora, aplicaremos a escala robusta aos nossos dados. Vamos usar a classe `RobustScaler` do Scikit Learn.

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Carregar o conjunto de dados Iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Criar uma figura com dois subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plotar os dados originais
axs[0].scatter(df['sepal length (cm)'], df['sepal width (cm)'], c='blue', label='Original Data')
axs[0].set_xlabel('Sepal Length (cm)')
axs[0].set_ylabel('Sepal Width (cm)')
axs[0].set_title('Dados Originais')

# Aplicar a escala robusta aos dados
scaler = RobustScaler()
df_robust_scaled = scaler.fit_transform(df)

# Criar um DataFrame com os dados escalados
df_scaled = pd.DataFrame(df_robust_scaled, columns=df.columns)

# Plotar os dados escalados
axs[1].scatter(df_scaled['sepal length (cm)'], df_scaled['sepal width (cm)'], c='red', label='Scaled Data')
axs[1].set_xlabel('Sepal Length (Scaled)')
axs[1].set_ylabel('Sepal Width (Scaled)')
axs[1].set_title('Dados Escalados com Robust Scaling')

# Mostrar o gráfico
plt.legend()
plt.tight_layout()
plt.show()
```

Benefícios da Escala Robusta
----------------------------

A principal vantagem da escala robusta é sua resistência a outliers. Ela utiliza a mediana e o intervalo interquartil (IQR) em vez da média e do desvio padrão, o que a torna mais apropriada para dados com valores extremos. Isso é particularmente útil em aplicações do mundo real, onde os outliers são comu A escala robusta é uma técnica valiosa quando precisamos dimensionar dados que podem conter outliers. Ela nos ajuda a manter a integridade dos dados, permitindo que nossos modelos de machine learning sejam mais robustos e precisos em situações do mundo real.

##  Rules of Thumb

Aqui estão algumas regras práticas (Rules of Thumb) para ter em mente ao realizar o dimensionamento (scaling) de recursos (features) em seus dados:

1. **Padronização (Standardization):**
    
    * Use a padronização quando você tiver recursos com diferentes escalas e desejar que eles tenham média zero e desvio padrão igual a um.
    * É particularmente útil quando você está trabalhando com algoritmos sensíveis à escala, como Support Vector Machines (SVM) e métodos baseados em gradiente.
2. **Normalização (Normalization):**
    
    * Utilize a normalização quando você quiser que os recursos estejam no intervalo [0, 1].
    * É útil quando você deseja manter as relações proporcionais entre os valores dos recursos.
3. **Robust Scaling (Escala Robusta):**
    
    * Considere a escala robusta quando seus dados contiverem valores discrepantes (outliers).
    * Ela é resistente a valores discrepantes e pode ser uma boa escolha em cenários onde a distribuição dos dados não é normal.
4. **MinMax Scaling (Escala Min-Max):**
    
    * Use a escala Min-Max quando você quiser que seus dados estejam em um intervalo específico, diferente de [0, 1].
    * Esta escala permite que você ajuste os dados para qualquer intervalo desejado.
5. **Lembre-se da Importância do Contexto:**
    
    * Não existe uma abordagem de dimensionamento única que funcione para todos os cenários. A escolha entre as técnicas de escala depende do contexto do seu problema e da natureza dos seus dados.
6. **Experimente Diferentes Abordagens:**
    
    * Não hesite em experimentar várias técnicas de escala e avaliar o desempenho do seu modelo com cada uma delas. A escolha certa pode fazer uma diferença significativa.
7. **Mantenha o Controle:**
    
    * Sempre mantenha o controle sobre o que cada técnica de escala está fazendo com seus dados. Verifique os resultados e compreenda como eles afetam seus recursos.
8. **Use a Escala Adequada para Cada Modelo:**
    
    * Diferentes modelos podem se beneficiar de diferentes técnicas de escala. Esteja ciente dos requisitos de escala de cada modelo.

Lembre-se de que a escolha da técnica de escala deve ser baseada no conhecimento do seu domínio e nas necessidades específicas do seu problema de aprendizado de máquina. A exploração e experimentação são cruciais para encontrar a abordagem mais adequada.
